```python
inp1=gps
inp1.reset_index(drop=True,inplace=True)
```

```python
inp1
```

```python
inp1.Reviews=np.log1p(inp1.Reviews)
inp1.Installs=np.log1p(inp1.Installs)
```

```python
inp1
```

```python
inp1.drop(columns=['App','Last Updated','Current Ver','Android Ver'],axis=1,inplace=True)
```

```python
inp1.shape
```

```python
inp2=pd.get_dummies(inp1,columns=['Category','Genres','Content Rating','Type'],drop_first=True)
inp2
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train,df_test=train_test_split(inp2,test_size=0.3,random_state=100)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
y_train=df_train.Rating
x_train=df_train.drop(['Rating'],axis=1)

y_test=df_test.Rating
x_test=df_test.drop(['Rating'],axis=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
l_model=LinearRegression()
l_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
train_predicted=l_model.predict(x_train)
r2_score(y_train,train_predicted)
```

```python
test_predicted=l_model.predict(x_test)
r2_score(y_test,test_predicted)
```
