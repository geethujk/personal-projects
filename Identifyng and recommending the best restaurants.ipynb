```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rest_data=pd.read_excel('data.xlsx')
rest_data
```

```python
type(rest_data)
```

```python
rest_data.shape
```

```python
rest_data.info()
```

```python
count_code=pd.read_excel('Country-Code.xlsx')
count_code
```

```python
type(count_code)
```

```python
rest_data=pd.merge(rest_data,count_code,on="Country Code",how="left")
rest_data
```

```python
type(rest_data)
```

```python
rest_data.isnull().sum()
```

```python
type(rest_data)
```

```python
rest_data.dropna(axis=0,subset=['Restaurant Name'],inplace=True)
rest_data.reset_index(drop=True,inplace=True)
```

```python
rest_data.isnull().sum()
```

```python
rest_data.columns=rest_data.columns.str.replace(' ','_')
```

```python
rest_data.duplicated().sum()
```

```python
rest_data['Cuisines'].fillna('Others',inplace=True)
```

```python
rest_data.isnull().sum()
```

```python
countrywise_rest=rest_data.groupby(['Country_Code','Country']).agg(count=('Restaurant_Name','count'))
countrywise_rest
```

```python
city_wise=rest_data.groupby(['Country','City']).agg(count=('Restaurant_Name','count'))
city_wise.sort_values(by='count',ascending=False)
                                                    
```

```python
national_presence=rest_data.groupby(['Restaurant_Name','Country']).agg(count=('Country','count'))
national_presence.sort_values(by='count',ascending=False)
```

```python
has_table_booking=rest_data.groupby('Has_Table_booking').Restaurant_ID.count()
has_table_booking
```

```python
ratio_of_booking=has_table_booking.Yes/has_table_booking.No
ratio_of_booking
```

```python
has_online_delivery=rest_data.groupby('Has_Online_delivery').Restaurant_ID.count()
has_online_delivery
```

```python
percent_of_online_delivery=(has_online_delivery.Yes/(has_online_delivery.Yes+has_online_delivery.No))*100
percent_of_online_delivery
```

```python
plt.pie(has_online_delivery,labels=("OD not allowed","OD allowed"),radius=1.5,autopct='%0.0f%%',explode=[0,0])
plt.show()
```

```python
OD_votes=rest_data.groupby('Has_Online_delivery').Votes.sum()
OD_votes
```

```python
diff_in_votes=OD_votes.No-OD_votes.Yes
diff_in_votes
```

```python
v_cuisines=rest_data['Cuisines'].apply(lambda x: pd.Series(x.split(',')))
v_cuisines.columns=['Cuisine1','Cuisine2','Cuisine3','Cuisine4','Cuisine5','Cuisine6','Cuisine7','Cuisine8']
v_cuisines
```

```python
cuisine1_no=pd.DataFrame(v_cuisines['Cuisine1'].value_counts()).reset_index()
cuisine1_no=cuisine1_no.rename(columns={'index':'Cuisines','Cuisine1':'count'})
cuisine1_no
```

```python
cuisine2_no=pd.DataFrame(v_cuisines['Cuisine2'].value_counts()).reset_index()
cuisine2_no=cuisine2_no.rename(columns={'index':'Cuisines','Cuisine2':'count'})
cuisine2_no
```

```python
cuisine3_no=pd.DataFrame(v_cuisines['Cuisine3'].value_counts()).reset_index()
cuisine3_no=cuisine3_no.rename(columns={'index':'Cuisines','Cuisine3':'count'})
cuisine3_no
```

```python
cuisine4_no=pd.DataFrame(v_cuisines['Cuisine4'].value_counts()).reset_index()
cuisine4_no=cuisine4_no.rename(columns={'index':'Cuisines','Cuisine4':'count'})
cuisine4_no
```

```python
cuisine5_no=pd.DataFrame(v_cuisines['Cuisine5'].value_counts()).reset_index()
cuisine5_no=cuisine5_no.rename(columns={'index':'Cuisines','Cuisine5':'count'})
cuisine5_no
```

```python
cuisine6_no=pd.DataFrame(v_cuisines['Cuisine6'].value_counts()).reset_index()
cuisine6_no=cuisine6_no.rename(columns={'index':'Cuisines','Cuisine6':'count'})
cuisine6_no
```

```python
cuisine7_no=pd.DataFrame(v_cuisines['Cuisine7'].value_counts()).reset_index()
cuisine7_no=cuisine7_no.rename(columns={'index':'Cuisines','Cuisine7':'count'})
cuisine7_no
```

```python
cuisine8_no=pd.DataFrame(v_cuisines['Cuisine8'].value_counts()).reset_index()
cuisine8_no=cuisine8_no.rename(columns={'index':'Cuisines','Cuisine8':'count'})
cuisine8_no
```

```python
cuisine_count=pd.concat([cuisine1_no,cuisine2_no,cuisine3_no,cuisine4_no,cuisine5_no,cuisine6_no,cuisine7_no,cuisine8_no])
cuisine_count
```

```python
top_n=cuisine_count.groupby('Cuisines').sum()
top_n_restaurants=top_n.sort_values(by='count',ascending=False)
top_n_restaurants
```

```python
cuisine_count_byrest=rest_data.groupby('Restaurant_Name').Cuisines.count().reset_index()
cuisine_count_byrest.sort_values(by='Cuisines',ascending=False)
```

```python
dist_cost_currency = pd.DataFrame(rest_data.groupby(['Restaurant_Name','Currency']).agg(count=('Average_Cost_for_two','sum'))).reset_index()
dist_cost_currency.sort_values(by='count',ascending=False)
```

```python
rest_rating_distrubution=pd.DataFrame(rest_data.groupby('Aggregate_rating').agg(count=('Restaurant_Name','count'))).reset_index()
rest_rating_distrubution.sort_values(by='Aggregate_rating',ascending=False)
```

```python
rating_country_count=pd.DataFrame(rest_data.groupby(['Country','Aggregate_rating']).agg(count=('Restaurant_Name','count'))).reset_index()
rating_country_count.sort_values(by='Aggregate_rating',ascending=False)
```

```python
rating_factors=rest_data[['Restaurant_Name','Restaurant_ID','Country','City','Aggregate_rating','Average_Cost_for_two','Has_Table_booking','Has_Online_delivery','Price_range','Votes']]
```

```python
dummy=['Has_Table_booking','Has_Online_delivery']
rating_factors=pd.get_dummies(rating_factors,columns=dummy,drop_first=True)
rating_factors
```

```python
rating_factors=rating_factors.merge(cuisine_count_byrest,on='Restaurant_Name',how='left')
rating_factors.head()
```

```python
sns.regplot(x='Average_Cost_for_two',y='Aggregate_rating',data=rating_factors,marker='*')
```
